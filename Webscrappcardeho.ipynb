```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
chrome_options = Options()
chrome_options.add_argument("--start-maximized")  # Opens browser in full screen
chrome_options.add_argument("--disable-blink-features=AutomationControlled")  # Avoids detection

driver = webdriver.Chrome(options=chrome_options)  # Initialize browser
driver.get('https://www.cardekho.com/used-cars+in+bangalore')
time.sleep(5)  # Wait for page to fully load

last_height = driver.execute_script("return document.body.scrollHeight")  # Get initial height

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # Scroll to bottom
    time.sleep(5)  # Wait for new data to load

    new_height = driver.execute_script("return document.body.scrollHeight")  # Get new height

    if new_height == last_height:  # If no new content, break loop
        break
    last_height = new_height

cars = driver.find_elements(By.XPATH, "//div[@class='gsc_col-xs-12 gsc_col-sm-6 gsc_col-md-4 cardColumn']")

structured_data = []

for car in cars:
    try:
        details = car.text.split("\n")  # Split details by line breaks
        
        # Extract Year and Car Name
        year_and_name = details[0].split(" ", 1)
        year = year_and_name[0]
        car_name = year_and_name[1] if len(year_and_name) > 1 else "N/A"

        # Extract Mileage, Fuel Type, Transmission
        mileage = fuel_type = transmission = ""
        if len(details) > 1 and "•" in details[1]:
            specs = [x.strip() for x in details[1].split("•")]
            mileage = specs[0].replace("kms", "").strip() if len(specs) > 0 else ""
            fuel_type = specs[1] if len(specs) > 1 else ""
            transmission = specs[2] if len(specs) > 2 else ""

        # Extract Price
        price = details[2].replace("₹", "").replace("Lakh", "").strip() if len(details) > 2 else "N/A"

        # Extract Location
        location = details[-1] if len(details) > 3 else "N/A"

        # Append extracted data
        structured_data.append({
            "Year": year,
            "Car Name": car_name,
            "Mileage (kms)": mileage,
            "Fuel Type": fuel_type,
            "Transmission": transmission,
            "Price (Lakh)": price,
            "Location": location
        })

    except Exception as e:
        print(f"Error processing car details: {details}\nError: {e}")

driver.quit()  # Close browser

# Convert to DataFrame
car_df = pd.DataFrame(structured_data)

# Save as CSV
car_df.to_csv('bengalure_car_details.csv', index=False)

# Display DataFrame & Count
print(car_df)

```
