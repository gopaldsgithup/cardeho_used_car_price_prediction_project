```python
def convert_to_integer(value):
    try:
        # Handle NaN or empty values
        if pd.isnull(value) or str(value).strip() == '':
            return None

        # Convert to string to safely use .replace()
        value = str(value).replace('₹', '').strip()

        # Handle Crore
        if 'Crore' in value:
            value = value.replace('Crore', '').strip()
            value = float(value) * 10000000

        # Handle commas
        elif ',' in value:
            value = value.replace(',', '')
            value = float(value)

        # Handle Lakh or direct numbers
        else:
            value = float(value) * 100000

        return int(value)

    except ValueError:
        # In case of unexpected format
        return None

# ✅ Apply the function again
df["Price (Lakh)"] = df["Price (Lakh)"].apply(convert_to_integer)

# 🔍 View the DataFrame
df.head()
```

```python
def city(value):
  if "Ahmedabad" in value:
    return "Ahmedabad"
  elif "Bangalore" in value:
    return "Bangalore"
  elif "Chennai" in value:
    return "Chennai"
  elif "Delhi" in value:
    return "Delhi"
  elif "Hyderabad" in value:
    return "Hyderabad"
  elif "Kolkata" in value:
    return "Kolkata"
  elif "Mumbai" in value:
    return "Mumbai"

df["Location"]=df["Location"].apply(city)
```

```python
#hypotheses

continues=['Car Name', 'Fuel Type', 'Transmission', 'Location', 'Year']
categories=['Mileage (kms)', 'Price (Lakh)']


from scipy import stats

def two_sample(d1,d2):
  t=0
  f=0
  for i in  range(31):
    sample1=d1.sample(frac=0.03)
    sample2=d2.sample(frac=0.03)
    t_test,p_value=stats.ttest_ind(sample1,sample2)
    if p_value < 0.055:
      f=f+1
    else:
      t=t+1
  if t>f:
    return True
  else:
    return False

#defining function for categories vs categories
def chisqare_cat_vs_cat(d1,d2):
  return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.055 else False

def annova_test(d1,d2):
  group=df[d2].unique()
  data={}
  for i in group:
    data[i]=df[d1][df[d2]==i]
  f_value,p_value=stats.f_oneway(*[i for i in data.values()])  #task for baby
  if p_value < 0.055:
    return False
  else:
    return True

final={}
for i in df.columns:
  final[i]={}
  for j in df.columns:
    if (i in continues) and (j in continues):
      result=two_sample(df[i],df[j])
    elif  (i in continues) and (j in categories):
      result=annova_test(i,j)
    elif (i in categories) and (j in continues):
      result=annova_test(j,i)
    elif (i in categories) and (j in categories):
      result=chisqare_cat_vs_cat(df[i],df[j])
    if result:
      final[i][j]=1
    else:
      final[i][j]=0
     
```
